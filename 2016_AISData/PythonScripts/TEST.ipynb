```python
import os
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
folder = '/home/mapper/VGI_Data/2016_AISData/ProcessedData/Port_AIS/Charleston/'
port = 'Charleston'
r = []
for root, dirs, files in os.walk(folder):
    for name in files:
       r.append(os.path.join(root, name))
```

```python
allDF = pd.DataFrame()
for file in r:
    df = pd.read_csv(file)
    allDF = allDF.append(df)
    
outpath = folder + port 
allDF.to_csv(outpath + 'ALL2016.csv')
```

```python
geometry = gpd.GeoSeries(Point(xy) for xy in zip(allDF.LON, allDF.LAT))
gdf = gpd.GeoDataFrame(allDF, geometry=geometry)
gdf.crs = {'init': 'epsg:4326'}
gdf.to_file(outpath + '_ALL2016.shp')
```
